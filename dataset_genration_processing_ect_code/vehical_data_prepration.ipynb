```python
import os
import shutil

folder_path = r"D:\reseacr dataset\vehical"
files = np.array(os.listdir(folder_path))


light_destination = os.path.join(folder_path, "light")
mediam_destination = os.path.join(folder_path, "mediam")
heavy_destination = os.path.join(folder_path, "heavy")

os.makedirs(light_destination, exist_ok=True)
os.makedirs(mediam_destination, exist_ok=True)
os.makedirs(heavy_destination, exist_ok=True)

light_vehicles = ["Bicycle", "Rickshaw", "Wheelbarrow"]
mediam_vehicles = ["Bike", "Car", "CNG", "Easy-bike", "Horse-cart", "Leguna", "Van"]
heavy_vehicles = ["Bus", "Truck"]



for fil in files:
    if fil in light_vehicles:
        path = os.path.join(folder_path, fil)
        
        if os.path.isdir(path):
            for file in os.listdir(path):
                source = os.path.join(path, file)
                destination = os.path.join(light_destination, file)
                
                # ✅ Check if file already exists
                if not os.path.exists(destination):
                    shutil.move(source, destination)
                    print(f"✓ Moved {file} to light")
                else:
                    print(f"⚠ Skipped {file} (already exists)")
    
    elif fil in mediam_vehicles:
        path = os.path.join(folder_path, fil)
        
        if os.path.isdir(path):
            for file in os.listdir(path):
                source = os.path.join(path, file)
                destination = os.path.join(mediam_destination, file)
                
                if not os.path.exists(destination):
                    shutil.move(source, destination)
                    print(f"✓ Moved {file} to mediam")
                else:
                    print(f"⚠ Skipped {file} (already exists)")
    
    elif fil in heavy_vehicles:
        path = os.path.join(folder_path, fil)
        
        if os.path.isdir(path):
            for file in os.listdir(path):
                source = os.path.join(path, file)
                destination = os.path.join(heavy_destination, file)
                
                if not os.path.exists(destination):
                    shutil.move(source, destination)
                    print(f"✓ Moved {file} to heavy")
                else:
                    print(f"⚠ Skipped {file} (already exists)")

print("✅ Done!")
```

```python
import os
import shutil

folder_path = 'D:\\reseacr dataset\\vehical'
mediam_destination = os.path.join(folder_path, "mediam")
light_destination = os.path.join(folder_path, "light")
heavy_destination = os.path.join(folder_path, "heavy")

# Create destination folders
os.makedirs(mediam_destination, exist_ok=True)
os.makedirs(light_destination, exist_ok=True)
os.makedirs(heavy_destination, exist_ok=True)

# Define vehicle categories
light_vehicles = ["Bicycle", "Rickshaw", "Wheelbarrow"]
mediam_vehicles = ["Bike", "Car", "CNG", "Easy-bike", "Horse-cart", "Leguna", "Van"]
heavy_vehicles = ["Bus", "Truck"]

# Counters for each category
light_counter = 0
mediam_counter = 0
heavy_counter = 0

files = np.array(os.listdir(folder_path))

for vehicle_type in files:
    vehicle_path = os.path.join(folder_path, vehicle_type)
    
    # Skip if not a folder
    if not os.path.isdir(vehicle_path):
        continue
    
    # Process light vehicles
    if vehicle_type in light_vehicles:
        for image_file in os.listdir(vehicle_path):
            source = os.path.join(vehicle_path, image_file)
            
            # Get file extension
            name, ext = os.path.splitext(image_file)
            
            # ✅ Rename with counter: image_0.jpg, image_1.jpg, etc.
            new_filename = f"image_{light_counter}{ext}"
            destination = os.path.join(light_destination, new_filename)
            
            shutil.move(source, destination)
            print(f"✓ Light: {image_file} → {new_filename}")
            
            light_counter += 1
    
    # Process mediam vehicles
    elif vehicle_type in mediam_vehicles:
        for image_file in os.listdir(vehicle_path):
            source = os.path.join(vehicle_path, image_file)
            
            name, ext = os.path.splitext(image_file)
            new_filename = f"image_{mediam_counter}{ext}"
            destination = os.path.join(mediam_destination, new_filename)
            
            shutil.move(source, destination)
            print(f"✓ Mediam: {image_file} → {new_filename}")
            
            mediam_counter += 1
    
    # Process heavy vehicles
    elif vehicle_type in heavy_vehicles:
        for image_file in os.listdir(vehicle_path):
            source = os.path.join(vehicle_path, image_file)
            
            name, ext = os.path.splitext(image_file)
            new_filename = f"image_{heavy_counter}{ext}"
            destination = os.path.join(heavy_destination, new_filename)
            
            shutil.move(source, destination)
            print(f"✓ Heavy: {image_file} → {new_filename}")
            
            heavy_counter += 1

print(f"\n✅ Complete!")
print(f"Light images: {light_counter}")
print(f"Mediam images: {mediam_counter}")
print(f"Heavy images: {heavy_counter}")
```
