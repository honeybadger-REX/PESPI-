```python
import numpy as np

# 1. Define the total damage capacity (Design Life)
# For example: 5000 units represents the total life of the pavement
total_capacity = 5000 

cumulative_damage = 0
running_sum_of_rates = 0
count = 0
rul_results = []

# Assuming year_data has your damage and the rate calculated previously
for annual_dmg, rate in zip(year_data["annual_damage"], year_data["Annual damage rate"]):
    cumulative_damage += annual_dmg
    running_sum_of_rates += rate
    count += 1
    
    # Calculate the average rate up to this year
    avg_rate = running_sum_of_rates / count
    
    # Calculate RUL: (Health Left) / (Damage per Year)
    health_left = total_capacity - cumulative_damage
    
    # Prevent negative RUL if the road has already "failed"
    current_rul = max(0, health_left / avg_rate)
    
    rul_results.append(current_rul)

year_data["Predicted_RUL"] = np.array(rul_results)
```
